```python
import pandas as pd
```

```python
df = pd.read_csv('데이터셋/광주 인공지능 사관학교 개발자 ONE DAY(응답) - 설문지 응답 시트1.csv')
```

```python
df.columns = ['Time','Vote']
```

```python
df
```

```python
df.info()
```

```python
list_name = ['친해져죠', '운동하죠', '풍요롭죠', '욜로하죠']
list_count = [0, 0, 0, 0]

for i in range(28):
    if '친해' in df.loc[i,'Vote'] or 'C' in df.loc[i,'Vote'] :
        list_count[0] +=1
    if '운동' in df.loc[i,'Vote'] :
        list_count[1] +=1        
    if '풍요' in df.loc[i,'Vote'] :
        list_count[2] +=1       
    if '욜로' in df.loc[i,'Vote'] or 'E' in df.loc[i,'Vote'] :
        list_count[3] +=1      

for i in range(4):
    print('{0}의 팀은 {1}의표를 받았습니다.'.format(list_name[i],list_count[i]))
```
